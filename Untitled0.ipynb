```python
import numpy as np
from scipy.fft import rfft, irfft
from scipy.io import wavfile
from matplotlib import pyplot as plt
from google.colab import drive
from IPython.display import Audio, display
drive.mount('/content/drive')


sampling_rate, audio_data = wavfile.read('/content/drive/MyDrive/fart-8.wav')
display(Audio(audio_data, rate = sampling_rate, autoplay = True))
fft_data = rfft(audio_data)
plt.plot(fft_data)

num_freq_bins = len(fft_data)

cutoff_frequency = 10000 # Hz
cutoff_bin = int(cutoff_frequency / (sampling_rate / num_freq_bins))
fft_data[:cutoff_bin] = 0
plt.plot(fft_data)
plt.show()

filtered_audio_data = irfft(fft_data)

plt.plot(audio_data)
plt.plot(filtered_audio_data)

plt.show()
filtered_audio_data = filtered_audio_data[:len(audio_data)]
filtered_audio_data = np.int16(filtered_audio_data / np.max(np.abs(filtered_audio_data)) * 32767)
display(Audio(filtered_audio_data, rate = sampling_rate, autoplay = True))
wavfile.write('/content/drive/MyDrive/OUT.wav', sampling_rate, filtered_audio_data)
```
